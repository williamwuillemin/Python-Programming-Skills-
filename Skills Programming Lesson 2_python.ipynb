```python
# Task 1
try:                                                                                    # initial safety net for wrong input
    x = float(input("Please could you give me a number (it can contains digits): "))    # define the inital input
    if x < 100:                                                                         # fix the 100 benchmark (evaluate if below)
        print("Your number is smaller than 100")                                        # return message if condition fullfield
    else:                                                                               # we know from conditional else that it is bigger than 100
        print("Your number is higher than 100")                                         # return message
except ValueError:                                                                      # second part of the safety net (handle invalid input)
    print("Error - you have to enter a float number")                                   # error message
```

```python
# Other versions
try:                                                                                          # safety net
    number = float(input("Please could you give me a number (it can contains digits): "))     # input form user transform to float
    if number < 100:                                                                          # condition smaller than 100
        print("Your number is smaller than 100")                                              # message if condition fullfilled
    elif number == 100:                                                                       # addition of small case when it is equal to zero (mainly for informative purpose, not needed in the instruction)
        print("Your number is equal to 100")                                                  # return message
    else:                                                                                     # all other case (stricly bigger than 100)
        print("Your number is higher than 100")                                               # return message
except ValueError:                                                                            # second part of safety net to handle wrong input
    print("Error - you have to enter a float number")                                         # error message
```

```python
# Task 2
try:                                                                            # safety net
  c = float(input("Please could you give me a temperature in Fahrenheit: "))    # Initial temperature input given by user
  f = (c - 32) * 5 / 9                                                          # Conversion formula
  print(f"The temperature in degree Celcius is {f}.")                           # result
except ValueError:                                                              # second part of safety net to handle wrong input
  print("Error, could you please enter a valid number")                         # Error message in case string instead of number
```

```python
# Other version
try:                                                                                          # safety net (invalid input)
    temperature = float(input("Enter the temperature: "))                                     # input termperature by the user
    choice = input("To convert to Celsius put a C or to Fahrenheit with an F? ").upper()      # selection of conversion by the user
    # conditional structure depending on temperature conversion
    if choice == "C":                                                                         # condition for Celsius (choice from the user)
        result_a = (temperature - 32) * 5 / 9                                                 # formula in one direction (C to F)
        print(f"The termperature {temperature} in Fahrenheit equals {result_a} in Celicus")   # print result
    elif choice == "F":                                                                       # condition for Fahrenheit
        result_b = (temperature * 9 / 5) + 32                                                 # formula in the other direction (F to C)
        print(f"The temperature {temperature}°C equal {result_b} in Fahrenheit")              # print result
    else:                                                                                     #  invalid choice of conversion
        print("Invalid choice. Please enter C or F.")                                         # error message for conversion choice
except ValueError:                                                                            # second part of safety net to handle wrong input (for the temperature)
    print("Error: please enter a valid number.")                                              # error message
```

```python
# Task 5 corrections
# safety net
try:
    # input (surname and name)
    vorname, nachname = input("Dear user, could you give me your first name and your surname (separated by a comma). Only alphabetical characters are accepted: ").strip().split(",")
    # need to remove the spacing otherwise it return an error like with William, Wuillemin instead of William,Wuillemin
    # check for correct style of name (dont contain number of strange sign like $)
    if vorname.isalpha() and nachname.isalpha():                                                            # check for letter only
        print("Perfect, your name only contains letters.")                                                  # print validation message
        # this second part check if the name is correct and ask for the validation of the user
        vor_x = input(f"Is your first name: {vorname}? Please, reply by Yes or No: ").upper()               # validation name (note it does not matter how you write it the program will return YES or NO in capital lettter)
        nach_x = input(f"Is your last name: {nachname}? Please, reply by Yes or No: ").upper()
        # check for only the negative case (if it is correct no need to change) -> != is for not YES
        if vor_x != "YES":                                                                                  # in case it is a yes it validate the name
            vorname = input("Could you write your correct first name: ")
        if nach_x != "YES":                                                                                 # in case it is a no it askes again for the name
            nachname = input("Could you write your correct last name: ")
        # Now that we have the correct name we can print output
        if vorname.isalpha() and nachname.isalpha():                                                        # check again that the name is only composed of letter
          print(f"Hello, {vorname} {nachname}.")                                                            # this time print with the correction
        else:                                                                                               # if there is once again a mistake
          print("Invalid, it can only contain alphabetical characters.")                                    # print invalid message due to issue with the name (not only letter)
    else:
        print("Invalid, it can only contain alphabetical characters.")                                      # first case where the name did not only contain letter the first time
# if people only give a first name and not last name like with William instead of William,Wuillemin
except ValueError:                                                                                          # error message if the user does not use a comma to separate the entry
    print("Error, could you please enter a valid input (two names separated by a comma).")
```

```python
# Task 5 improvement with a while loop until the to check the value until the user validate it.

while True:               # create the infinit loop (note that it is risky and will need break everywhere to make sure that it stop)
    try:                  # safety net if the input is not valid and make the program crash
        vorname, nachname = input("Give me your first name and surname (separated by a comma): ").strip().split(",")    # input separated by the comma (remove spacing to avoid formating error)
        if not vorname.isalpha() or not nachname.isalpha():                                                             # check if the name only contain letter
            print("Invalid, it can only contain alphabetical characters.")                                              # return error message if it is not the case
            continue                                                                                                    # this time the loop continue in order to give the opportunity to the user to change it.
        vor_x = input(f"Is your first name: {vorname}? (Yes/No): ").upper()                                             # check the validity of the name (YES/NO but also validate yes and no, due to upper (transform in upper case))
        if vor_x != "YES":                                                                                              # if it is not yes, the user can change it
            vorname = input("Could you write your correct first name: ").strip()                                        # new input for the name from the user

        nach_x = input(f"Is your last name: {nachname}? (Yes/No): ").upper()                                            # same system but for the sirname this time
        if nach_x != "YES":
            nachname = input("Could you write your correct last name: ").strip()

        if vorname.isalpha() and nachname.isalpha():                                                                    # once it return yes to both then the program print the result and break the while loop
            print(f"Hello, {vorname} {nachname}.")                                                                      # print the name
            break                                                                                                       # end of the loop
        else:                                                                                                           # the inital case where it is not only letter
            print("Invalid, it can only contain alphabetical characters.")                                              # error message

    except ValueError:                                                                                                  # handle error value with error link to the formating of the first reply
        print("Error, please enter two names separated by a comma.")
```
