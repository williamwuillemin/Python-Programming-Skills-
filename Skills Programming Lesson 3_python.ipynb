Skills Programming Lesson 3 - Conditionals

```python
# Task 1
try:
  x = input("Please give me two numbers (separate them by a comma): ").split(",")     # create a list with the two element
  a, b = float(x[0]), float(x[1])                                                     # make sure we can also give a float and not only an integer
  if a == b:                                                                          # the case where they are equal
    print("The two numbers given are equal: %f" %a)
  else:                                                                               # print the smallest value
    print(f"This is the smallest value given {min(a,b)}")
except ValueError:
  print("Please give me a valid value")                                               # take care if someone put a letter
```

```python
# Task 2
try:
    a, b = [int(x) for x in input("Enter two non-zero integers (separate by comma): ").split(",")]        # same system as in previous task but this time try with a for loop to split the list of two value
    if a == 0 and b == 0:                                                                                 # analse the case with two 0 input
      print("Please enter two non-zero integers")
    else:
      if ((a > 0) and (b < 0)) or ((a < 0) and (b > 0)):                                                  # if for the two case and otherwise it goes to else and print No
        print("YES")
      else:
        print("NO")
except ValueError:                                                                                        # just take care of potential letter instead of number as an input
    print("Please enter valid integers.")
```
