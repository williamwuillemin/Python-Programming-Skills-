```python
# Task 1
try:                                                                                  # safety net start
  x = input("Please give me two numbers (separate them by a comma): ").split(",")     # create a list with the two element with input from user
  a, b = float(x[0]), float(x[1])                                                     # make sure we can also give a float and not only an integer for each of the element of the list
  if a == b:                                                                          # the case where they are equal
    print("The two numbers given are equal: %f" %a)                                   # result in equality message
  else:                                                                               # print the smallest value
    print(f"This is the smallest value given {min(a,b)}")                             # return the smallest value of the two
except ValueError:                                                                    # error value if the input is a letter or wrong format
  print("Please give me a valid value")                                               # error message if someone put a letter
```

```python
# Task 2
try:                                                                                                      # safety net
    a, b = [int(x) for x in input("Enter two non-zero integers (separate by comma): ").split(",")]        # same system as in previous task but this time try with a for loop to split the list of two value (could also use map)
    if a == 0 and b == 0:                                                                                 # analse the case with two 0 input
      print("Please enter two non-zero integers")                                                         # return message for zero
    else:                                                                                                 # conditional for the remaining case
      if ((a > 0) and (b < 0)) or ((a < 0) and (b > 0)):                                                  # if for the two case and otherwise it goes to else and print No
        print("YES")                                                                                      # yes if one of the two is positiv
      else:
        print("NO")                                                                                       # no in the remaining case (in teh case where they are both positiv)
except ValueError:                                                                                        # second side of the  safety net, just take care of potential letter instead of number as an input that would make the program crash
    print("Please enter valid integers.")                                                                 # message regarding wrong imput
```
