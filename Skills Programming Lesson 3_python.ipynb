Skills Programming Lesson 3 - Conditionals

```python
# Task 1
try:
  x = input("Please give me two numbers (separate them by a comma): ").split(",")     # create a list with the two element
  a, b = float(x[0]), float(x[1])                                                     # make sure we can also give a float and not only an integer
  if a == b:                                                                          # the case where they are equal
    print("The two numbers given are equal: %f" %a)
  else:                                                                               # print the smallest value
    print(f"This is the smallest value given {min(a,b)}")
except ValueError:
  print("Please give me a valid value")                                               # take care if someone put a letter
```
