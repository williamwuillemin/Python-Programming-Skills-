```python
# Task 1
import array                                                                                                # import the packages
try:                                                                                                        # safety net - prepare in case someone put letter or technical sign $
  # create the array with the input of user (transform every entry into an integer)
  my_array = array.array("i", [int(x) for x in input("Enter a list of integer to put in the initial array. Note that it needs to be separated by a comma").split(",")])
  print(f"The array is: {my_array}")                                                                        # print the result in order to check the array and see which element one want to delete
  value = int(input("Which element number do you want to delete? Please indicate its position in the order: "))                                       # input from user to delete an element
  if value < 1 or value > len(my_array):                                                                    # just add the conditional to make sure the element is in the array (note that we ask for the coordinate)
      print(f"Error: element number {value} does not exist in the array (valid range: 1-{len(my_array)})")  # return error message
  else:                                                                                                     # the right value can be delete as it is valid (it has a valid coordinate)
      del my_array[value - 1]                                                                               # delete the value depending on its coordinate (hence the -1 as the indexation start at -1)
      print(f"New array: {my_array} after the deletion of {value}")                                         # Result message
except ValueError:                                                                                          # second part of the safety net -prepare in case someone put letter or technical sign $
  print("Please give a valid input.")                                                                       # error message
```

```python
# Task 2
# the array only accept number

import array                                                                    # import the packages
try:                                                                            # safety net - prepare in case someone put letter or technical sign $
  # create the array with the input of user
  my_array = array.array("i", [int(x) for x in input("Enter a list of integer to put in the initial array. Note that it needs to be separated by a comma").split(",")])
  print(f"The array is: {my_array}")                                            # just check the array in order to see which element one want to delete
  value = int(input("Which element number do you want to add?: "))              # new element that the user want to add (input)
  if isinstance(value, int):                                                    # create the conditional part just to make sure the input is valid (only accept integer)
    my_array.append(value)                                                      # the true value that we just added
  else:
    print("Please give a valid input.")                                         # invalid input - if want to add a letter
  print(f"New array: {my_array} after the addition of {value}")                 # print the result after adding the new element
  print(f"The lenght of the array is : {len(my_array)}")
except ValueError:                                                              # prepare in case someone put letter or technical sign $
  print("Please give a valid input.")                                           # error message
```
