```python
# Task 1

# we are asked to print the date of today, yesterday and tomorrow:

# In order to do this there are two options we either do everything without a package and simply raw code everything with function or we can use packages

# version without packages (this resemble a bit what we did in one of the previous exercise where we had to give the date).

# we first need to check if it is a leap year because then it is going to change what are the end of the months.
def leap_year(y):
    return y % 4 == 0 and (y % 100 != 0 or y % 400 == 0)
    # return true if it is a leap year and false if it isn't (note that century are not leap year but the only exception is every 400, there is a leap year)
# now we can check for the correct end of the month depending on the leap year or not
def end_month(m, y):
    return [31, 29 if leap_year(y) else 28, 31, 30, 31, 30,
            31, 31, 30, 31, 30, 31][m - 1]
            # if it is a leap year, we know that feb is 29 days long and otherwise 28, the other one are the usual one
# function that fixes the date for yesterday and tomorrow
def add_days(y, m, d, minus):
    d += minus
    # we add a minus parameter to substract and add 1 day to find tomorrow and yesterday
    while d > end_month(m, y):
        d -= end_month(m, y)
        m += 1
        if m > 12:
            m = 1
            y += 1
            # we check if the number of day exceed the max after the addition of days if it does we first substract the max_month day to the number of day to find the day of next month: imagine using minus +1 with the date 31 -> we obtain 32 which is bigger than any month
            # thus we substract 32-31 = 1 -> the next day is going to be the first of next month
    while d < 1:
        m -= 1
        if m < 1:
            m = 12
            y -= 1
        d += end_month(m, y)
            # this is the same system but the other way around we substract -1 to the number of day, let say 1 - 1 which = 0 and is impossible so it move to the previous month and uses the max number of day of the previous day for the day (note that if it happen on the 1 of jan it goes one year back in time)
    return y, m, d
# print the final date in the right format
def date_final(y, m, d):
    return(f"{y}-{m:02d}-{d:02d}")
    # note that it add a 0 if shorter that 2 digit and the 2 indicates to only print 2 digit

try:
    # now we can create the concrete output
    y, m, d = map(int, input("Enter the date of today (2026-04-28): ").split("-"))
    if m < 1 or m > 12:
        print("Month must be between 1 and 12.")
    elif d < 1 or d > end_month(m, y):
        print(f"Day must be between 1 and {end_month(m, y)} for that month.")
        # just check that we have a realistic month or day for the month
    else:
        yesterday = add_days(y, m, d, -1)
        tomorrow  = add_days(y, m, d,  1)
        # apply the function to add and substract one day
        print(f"Yesterday : {date_final(*yesterday)}")
        print(f"Today     : {date_final(y, m, d)}")
        print(f"Tomorrow  : {date_final(*tomorrow)}")
        # note that we need to add * because otherwise the program see ((2026-04-28)) and print an error instead with * it sees (2026-04-28) and work
except ValueError:
    print("Invalid format. Please use YYYY-MM-DD with numbers only.")
    # just make sure that the date is in the right format

```
