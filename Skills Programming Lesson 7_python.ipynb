Skills Programming - Lesson 7 Loops

```python
# Task 1
try:
  a = input("Please enter numbers (separated by a comma): ").split(",")
  numbers = []
  for item in a:
    item = item.strip()
    numbers.append(int(item))
  print(f"The sum of your input is {sum(numbers)}.")
except ValueError:
  print("Please enter numbers separated by a comma")
```

```python
# Task 1 (different interpretation)
a = input("Please enter numbers (separated by a comma): ").split(",")
total = 0
for item in a:
  try:
    item = item.strip()
    total += int(item)
  except ValueError:
    print(f"We are skipping the entry {item} as it is not a valid one")

print(f"The total sum of the numbers given is {total}")
```

```python
# Task 1 (other options)
a = input("Please enter numbers (separated by a comma): ").split(",")                         # takes the list of input and make sure they are split (using the comma)
total = 0                                                                                     # create the inital sum of value in which we are going to store the additional
for item in a:                                                                                # the loop that goes through each element given as input one by one
    item = item.strip()                                                                       # need to add this because the program consider the input as an invalid input (remove the spacing)
    while True:                                                                               # start the loop to add item and break once this is done (note that if they are invalid it does into the expect ValueError)
        try:                                                                                  # this is the part that makes sure that all the input are valid, before transforming the string into integer and summing them
            total += int(item)
            break                                                                             # valid input, move to next item
        except ValueError:
            item = input(f"The input '{item}' is not valid. Please enter a valid input: ")    # invalid input are isolated and the program asks to the user to replace the input in order to have a new compelete list. Note that it can also skip the item simply using: print(f"We are skipping the entry {item} as it is not a valid one")

print(f"The total sum of the numbers given is {total}")                                       # return the final sum
```
