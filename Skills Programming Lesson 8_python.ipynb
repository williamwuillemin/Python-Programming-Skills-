```python
# Task 1
def sumthree():                                                                         # define the function
  try:                                                                                  # add in case someone type characters instead of numbers
    a,b,c = input("Enter three number separated by a comma: ").strip().split(",")       # takes the input (the tree numbers)
    numbers = []                                                                        # the create the list in which we are going to sort all the input once transform into float or int
    for i in [a,b,c]:                                                                   # selection loop (tranform all the input into either integer and float)
      if "." in i:                                                                      # float contain a dot
        numbers.append(float(i))                                                        # add input after being transform into a number
      else:                                                                             # character aleady excluded so only integer left
        numbers.append(int(i))                                                          # transform into an integer
  except ValueError:
    print("Please enter a valid input")                                                 # error message when the input is a character
    return                                                                              # stops here because otherwise it print 0 from result = 0
  result = 0                                                                            # define initial starting point for the addition
  for x in numbers:                                                                     # the loops that goes through all the numbers
    result += x                                                                         # sum all the component
  print(result)                                                                         # print the result

# invoke the function
sumthree()
```

```python
# Task 2
def myfunction():                                                                         # the definition of the function
  try:                                                                                    # protection against wrong input (characters and so on)
    a,b,c = input("Enter three number separated by a comma: ").strip().split(",")         # the input
    numbers = []                                                                          # list in which we are going to sort the value once transform
    for x in [a,b,c]:                                                                     # transform all the input into a float (so that we can also give the biggest value between 3.2 and 3.3)
      numbers.append(float(x))
    a,b,c = numbers                                                                       # all the value are now stored into the list numbers
    if a > b:                                                                             # we check the basic classification if a > b
      if a > c:                                                                           # if it is the case we need, now to see if a bigger than c
        print(a)                                                                          # as a is the biggest number we print a
      else:
        print(c)                                                                          # as c is the biggest number we print c
    else:                                                                                 # if we reach this part of the code it means that b is bigger than a
      if b > c:                                                                           # so we need to check if b is bigger than c
        print(b)                                                                          # if it is the case we return b
      else:                                                                               # if it is not the case we return c
        print(c)
  except ValueError:                                                                      # return an error message in case the input is a character or separeted by something else than a comma
    print("Please enter a valid input")

# invoke the function
myfunction()
```
