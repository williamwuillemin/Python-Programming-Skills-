# Python Skills Master - Data Handling


```python
# Task 1
#Task description
# in this task we want to understand the volume traded on the stock market
# in specific years by the predefined stocks.

import pandas as pd                     # pandas in order to work with the dataFrame
import datetime as dt                   # package for dates
import yfinance as yf                   # we need to import yfinance because it does not work from google as it does not exist anymore




# TASK 1:
# Create your time range (start and end variables).
# The start date should be 01/01/2015 and the end should today (whatever your today is)
start = dt.datetime(2015, 1, 1)               # we define from when we want to extract the data using the function datetime from the modul dt
end = dt.datetime.today()                     # we simply load the data until today



# we pick the Apple, Tesla, Twitter, IBM stocks symbols
# LNKD is not listed so it was changed to microsoft instead (MSFT instead - microsoft)
stocks = ['AAPL', 'TSLA', 'IBM', 'MSFT']

df = yf.download(stocks, start = start, end = end)          # use the function of the yf packages instead because google does not work (extract the data and add it to the dataset df)



vol = df['Volume'].copy()                                   # we now create a dataFrame called vol, with the Volume values. # need to add .copy() in order to avoid simply having a view of the colomn (we want the copy)
vol.head()                                                  # preview the first row


vol['week'] = vol.index.isocalendar().week                  # we now want to Aggregate the data of Volume to weekly (use isocalendar in order to indicate which week it is)
vol['year'] = vol.index.year                                # extract the year

week = vol.groupby(['week', 'year']).sum()                  # we group the data by weeks and year before doing the sum of everything (so all the day of the same week during the same year are sum up)
week.head()                                                 # preview the data




# TASK 2:                                                   # Find all the volume traded in the year of 2015
vol_2015 = vol[vol['year'] == 2015]                         # keep only 2015 rows
vol_2015 = vol_2015.drop('week', axis=1)                    # drop the week column (axis=1 means column)
result = vol_2015.groupby('year').sum()                     # group by year and sum everything

# Format the output in order to have a nice visual
result.columns = ['Apple', 'IBM', 'Microsoft', 'Tesla']     # create the list of all the company name
print("\nTotal Volume Traded in 2015")                      # simply add a title between the ***100%*** indicating in the printing (note that needed to add \n to jump a line otherwise everything is glude together)
for col in result.columns:                                  # create a loop that goes through each company in the list
    value = result[col].values[0]                           # extract the result for each company (each row is a company) and uses the first row (because it is the only one with the result)
    print(f"  {col:>12}:  {value:>20,.0f} shares")          # print the result (:>12 align everything 12 characters to the right, and it is the same for >20 afterwards, the .0f removes the decimal and ,print a comma every 3 digit)
```
